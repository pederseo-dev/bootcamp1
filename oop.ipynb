## Programación orientada a objetos

Python también nos permite representar cosas del mundo real, más fielmente por medio de estructuras de datos llamadas **objetos**.

Tanto sus atributos (características y propiedades) como sus métodos (funciones que nos permiten interactuar con el objeto) se definen en una `clase`.

Esta `clase` es como una plantilla para los objetos que queremos crear, como un molde.

Luego ese molde se utiliza para crear la cantidad que necesitemos de los `objetos` de esa `clase`.



```python
# ejemplo 1
# crear el molde
# class Torta:
#     harina = 200
#     huevo = 3
#     azucar = 200
#     manteca = 100
#     leche = 1

# creando el objeto
torta1 = Torta()
print('cantidad de harina', torta1.harina)

torta2 = Torta()
print('cantidad de huevos', torta2.huevo)
```

## Definiendo atributos
Es importante distinguir que existen dos tipos de atributos:

`Atributos de clase`: Son atributos cuyos valores definen a la clase, por lo tanto serán comunes para todos los objetos. (Mostrar el que ya teníamos arriba)

`Atributos de instancia`: Son atributos cuyos valores son específicos de los objetos creados (la instancia). En nuestro caso, de ambas mascotas tiene el atributo `nombre`, pero cada una tiene un valor diferente, que se define al crear el objeto. (Mostrar los de instancia)


```python
class Torta:
    harina = 200
    huevo = 3
    azucar = 200
    manteca = 100
    leche = 1

    def __init__(self, sabor, cobertura):
        self.sabor = sabor
        self.cobertura = cobertura


torta1 = Torta('chocolate', 'frutilla')
print(torta1.sabor)

torta2 =Torta('frutilla', 'vainilla')
print(torta2.sabor)
```

## 📌 ¿Qué es self?
self representa al propio objeto que se está creando o usando en ese momento. Es una convención que se utiliza dentro de los métodos de una clase para acceder a los atributos y otros métodos del objeto actual.

Es como decir:

“Esto pertenece a mí, el objeto que está siendo creado o que está ejecutando este método.”

## 👀 ¿Se puede cambiar la palabra self?
Técnicamente sí, se puede usar otro nombre en su lugar. Python solo necesita que el primer parámetro del método reciba una referencia al objeto actual, pero por convención siempre se usa self.

```python
class Penguin_academy:
    def __init__(self, nombre):
        self.nombre = nombre

pinguino = Penguin_academy('olaf')
print(pinguino.nombre)
```

Ejemplo: "Cuando me muevo, hago X".



```python
# solucion de alan
class Animales():
    especie = 'Ave'
    def __init__(self,color,nombre):
        self.color = color
        self.nombre = nombre

    def hablar (self,animal):
        print(f'El/la {animal} {self.nombre} esta hablando')

    def mover (self,animal):
        print(f'El/la {animal} {self.nombre} se esta moviendo')

pajaro = Animales('azul','juan')
print(f'Me llamo {pajaro.nombre} y soy una {pajaro.especie} de color {pajaro.color}')
pajaro.hablar(pajaro.especie)
pajaro.mover(pajaro.especie)
```

```python
class Animales:
    especie = 'vacas'

    def __init__(self, nombre, posicion=0):
        self.nombre = nombre
        self.posicion = posicion

    def correr(self, metros):
        self.metros = metros
        self.posicion = self.posicion + self.metros
        print(f'la vaca {self.nombre} corrio {self.metros}')

    def escapa(self):
        if self.posicion > 50:
            print(f'la vaca {self.nombre} logro escapar')
        else:
            print(f'la vaca {self.nombre} no pudo escapar :c')

vaca1 = Animales('lola')
print(vaca1.nombre)

vaca1.correr(40)
vaca1.escapa()
vaca1.correr(20)
vaca1.escapa()
```

## Herencia
Ojo! Esto no es una herencia como tal, sino una relación entre clases.
Lo que representa la herencia es que una clase hereda los atributos y métodos de otra clase. (mostrar ejemplo)

Dado que una clase hija hereda los atributos y métodos de la clase padre, nos puede ser muy útil cuando tengamos clases que se parecen entre sí pero tienen ciertas particularidades

```python
class Reptil:
    tipo = 'Reptil'

    def __init__(self, color, sonido):
        self.color =color
        self.sonido = sonido

    def hablar(self):
        print('buenos dias')

    def presentarse(self):
        print(f'soy un animal del tipo {self.tipo}')


class Serpiente(Reptil):
    veneno = 10

    def arrastrarse(self):
        print('Me estoy arrastrando')


class Jakare(Reptil):
    garras = 4

    def morder(self):
        print('Grrr te muerdo')


curiju = Serpiente('marron' ,'ssssss')
curiju.presentarse()
curiju.hablar()


```

## Challenge 🤺15 min
Crear dos personajes de una historia que se representen a traves de clases. Ejemplo: Un heroe y un villano.
Cada personaje debe tener poderes o acciones diferentes. Y deben presentarse en una frase.

Ejemplo: Yo me llamo pepito y tengo super fuerza
Ejemplo: Yo me llamo juanito y puedo volar

```python
class Dinosaurio:
    tipo = "Dinosaurio"

    def __init__(self, color, nombre, habilidad):
        self.color = color
        self.nombre = nombre
        self.habilidad = habilidad

    def presentación(self):
        print(f"Buenas!! Me llamo {self.nombre}.")

    def talento(self):
        print(f"Mi talento es {self.habilidad}!")

    def rugido(self):
        print(f"RAWRR >:o")

class Pterodactyl(Dinosaurio): 
    velocidad_vuelo = 70
    def volar(self):
        print(f"*vuela a {self.velocidad_vuelo}*")

class Velociraptor(Dinosaurio):
    velocidad_corrida = 40
    def correr(self):
        print(f"*corre a {self.velocidad_corrida}*")

blue = Velociraptor("azul", "Blue", "morder")
blue.presentación()
blue.talento()
blue.correr()

ave = Pterodactyl("rojo", "Ave", "cazar")
ave.presentación()
ave.rugido()
ave.talento()
ave.volar()
```
