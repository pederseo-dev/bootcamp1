### Dos condiciones a la vez - Opeardores lógicos

### `Operadores lógicos`
- `and` (y) - ambas condiciones deben ser verdaderas    
- `or` (o) - al menos una de las condiciones debe ser verdadera
- `not` (no) - invierte el valor de la condicion

```python
edad_laboral = 66
trabaja = False


if edad_laboral < 18:
    print('menor de edad')

elif edad_laboral < 75 and trabaja == True:
    print('activo laboralmente')

elif edad_laboral < 65 and trabaja == False:
    print('estudiante')

else:
    print('jubilado')
```

```python
def chequear_edad(edad_laboral, trabaja):
    if edad_laboral < 18:
        print('menor de edad')

    elif edad_laboral < 75 and trabaja == True:
        print('activo laboralmente')

    elif edad_laboral < 65 and trabaja == False:
        print('estudiante')

    else:
        print('jubilado')

chequear_edad(70, True)

```

### Bucles condicionales

While nos permite ejecutar una serie de instrucciones mientras una condicion se cumpla. Al momento que la condicion deje de ser cierta el bucle va parar.

- Ejemplo: Imagina que estas saltando la cuerda, mientras no te tropieces vas a seguir saltando, pero si te tropiezas el bucle se detiene.

```python
# ejemplo 1
numero_a_adivinar = 30
numero = 0

while numero != numero_a_adivinar:
    numero = int(input('ingrese un numero'))
    print(numero)

print('numero correcto')







```

# bucles infinitos
Un bucle infinito sucede cuando la condicion siempre es cierta y por lo tanto el bucle no termina de ejecutarse. "Break" es una palabra reservada que nos permite salir de un bucle.

```python
# ---------------------------ejemplo 1
numero_a_adivinar = 30
numero = 0

while numero != numero_a_adivinar:
    numero = int(input('ingrese un numero'))
    print(numero)

print('numero correcto')

# ---------------------------ejemplo 2
numero_adivinar = 1

while True:
    numero_igresado = int(input('ingrese un numero'))
    if numero_igresado == numero_adivinar:
        print('el numero es correcto')
        break

    else:
        print('el numero es incorrecto, intente de nuevo')
    
```

### Bucles iterativos

Cuando queremos repetir una acción una cierta cantidad de veces, podemos utilizar un bucle iterativo.

Iteraciones: una acción repetida una cierta cantidad de veces, pero en cada repetición se tiene en cuenta el valor anterior.

Para eso podríamos usar un generador de iterables llamado "range", de la siguiente forma:



```python
# ejemplo 1 
for i in range(10):
    print('hola, buen dia')

# ejemplo 2
for iterador in range(5):
    print(iterador)
```
