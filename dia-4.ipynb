## Diccionarios
Un diccionario es una colección de pares formados por una clave y un valor asociado a la clave. Se construyen poniendo los pares entre llaves `{ }` separados por comas, y separando la clave del valor con dos puntos :.

Se caracterizan por:

- No tienen orden.
- Pueden contener elementos de distintos tipos.
- Son mutables, es decir, pueden alterarse durante la ejecución de un programa.
- Las claves son únicas, es decir, no pueden repetirse en un mismo diccionario, y pueden ser de cualquier tipo de datos.

```python
# Ejemplo de diccionario
mi_contacto = {
    "nombre":"Analia", 
    "apellido":"Britez", 
    "mail":"analia@penguin.com", 
    "telefono":'0984334455',
    "direccion":"calle penguin 1",
    "frutas":"banana"
    }

# visualizar diccionario
# print(mi_contacto)

# print(mi_contacto["nombre"])
# print(mi_contacto["direccion"])

# ejemplo Desplegar datos del diccionario
# print(mi_contacto.keys())
# print(mi_contacto.values())
# print(mi_contacto.items())


# Ejemplo anhadir un elemento
print(mi_contacto)
mi_contacto["barrio"] = "los laureles"
print(mi_contacto)

# Ejemplo actualizar un dato
mi_contacto["barrio"] = "Recoleta"
print(mi_contacto)

# Ejemplo eliminar
mi_contacto.popitem()
print(mi_contacto)

# Ejemplo eliminar por clave
del mi_contacto["apellido"]
print(mi_contacto)
```

```python
tarjeta_personal={}

def crear():
    clave=input("Ingrese el nombre de la clave")
    valor=input("Ingrese un valor para la clave")
    tarjeta_personal[clave]=valor


def modificar():
    clave=input("Ingrese el nombre de la clave a modificar")
    valor=input("Ingrese un nuevo valor")
    tarjeta_personal[clave]=valor

def eliminar():
    clave= input("Ingrese el nombre de la clave que desea eliminar")
    del tarjeta_personal[clave]


while True:
    opcion=int(input("Ingrese una opcion: 1. Crear, 2. Modificar, 3. Eliminar, 4. Salir"))
    if opcion == 1:
        crear()
        print(tarjeta_personal)
    elif opcion ==2:
        modificar()
        print(tarjeta_personal)
    elif opcion==3:
        eliminar()
        print(tarjeta_personal)
    else :
        break
```
